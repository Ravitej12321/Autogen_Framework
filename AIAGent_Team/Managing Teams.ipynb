```python
from autogen_agentchat.teams import RoundRobinGroupChat
from autogen_agentchat.agents import AssistantAgent
from autogen_agentchat.messages import TextMessage
from dotenv import load_dotenv
from autogen_ext.models.openai import OpenAIChatCompletionClient
import os,asyncio
load_dotenv()
gemini_api_key = os.getenv("GEMINI_KEY")
model_client = OpenAIChatCompletionClient(
    api_key = gemini_api_key,
    model = "gemini-2.0-flash",
)

code_developer = AssistantAgent(
    name = "code_developer",
    model_client = model_client,
    description = "you are a excellent code developer who can write code for the given task on python",
    system_message = "you are a excellent code developer who can write code for the given task on python",
)

code_reviewer = AssistantAgent(
    name = "code_reviewer",
    model_client = model_client,
    description = "you are a excellent code reviewer who can review the code for the given task on python",
    system_message = "you are a excellent code reviewer who can review the code for the given task on python",
)
code_tester = AssistantAgent(
    name = "code_tester",
    model_client = model_client,
    description = "you are a excellent code tester who can test the code for the given task on python",
    system_message = "you are a excellent code tester who can test the code for the given code in all scenarios on python",
)





```

```python
from autogen_core import CancellationToken
from autogen_agentchat.ui import Console
cancellation_token = CancellationToken()
team = RoundRobinGroupChat(
    participants = [code_developer,code_reviewer,code_tester],
    max_turns = 3,
)
response = asyncio.create_task(
Console(team.run_stream( task = "write a python code for printing hello world",cancellation_token=cancellation_token))
)
await asyncio.sleep(2)
cancellation_token.cancel()

try : 
    await response
except :
    print("Task was End")
```

```python
async def single_agent_task():
    task = TextMessage(content = "Write a small village agent story in a sarcastic way.",source='user')
    response = await single_agent.on_messages(messages=[task],cancellation_token=CancellationToken())
    print(response)
    print(response.chat_message)
await single_agent_task()
```

```python
saved_state = await single_agent.save_state()
print(saved_state)
```

```python
from autogen_agentchat.agents import AssistantAgent
from autogen_ext.models.openai import OpenAIChatCompletionClient
from autogen_agentchat.messages import TextMessage
from autogen_core import CancellationToken
from autogen_agentchat.conditions import TextMentionTermination
from dotenv import load_dotenv
load_dotenv()
api_key = os.getenv("GEMINI_KEY")
print(api_key)
model_client = OpenAIChatCompletionClient(model="gemini-2.0-flash",api_key = api_key)
from autogen_agentchat.teams import RoundRobinGroupChat
first_agent =  AssistantAgent(
    name = "Agent_model_1",
    model_client = model_client,
    system_message = "You are a Stock News analyser under 50 words"
    
)
second_agent =  AssistantAgent(
    name = "Agent_model_2",
    model_client = model_client,
    system_message = "You are a Stock Market Trader under 15 min chart and signal buy/sell Once completed end with Text 'THE END'"
)
team = RoundRobinGroupChat(
    participants= [first_agent,second_agent],
    termination_condition=TextMentionTermination('THE END'),
    max_turns = 2
)
```

```python

response = team.run_stream(task = "Analyse the stock Reliance",cancellation_token=CancellationToken())

from autogen_agentchat.ui import Console
result = await Console(response)
print(result)
```

```python
team_state = await team.save_state()

print("team State has been saved.")
```
