# Structured Output


```python
import pydantic
from pydantic import BaseModel

class Chemistry(BaseModel):
    Element: str
    Scientific_Name: str
    Formula: str
    atomic_number: int
```

```python
from autogen_agentchat.ui import Console
from autogen_agentchat.messages import TextMessage
from dotenv import load_dotenv
from autogen_ext.models.openai import OpenAIChatCompletionClient
from autogen_core import CancellationToken
import os
load_dotenv()
api_key = os.getenv("GEMINI_KEY")
if not api_key:
    raise ValueError("GEMINI_KEY is not set")
model_client = OpenAIChatCompletionClient(
    api_key = api_key,
    model = "gemini-2.0-flash",
    response_format = Chemistry
)
```

```python
from autogen_agentchat.agents import AssistantAgent

assistant = AssistantAgent(
    name = "assistant",
    model_client = model_client,
    description = "you are a great text agent who can answer questions and help with tasks",
    system_message = "you are a chemistry scientist had a indepth knowledge on chemical elements",
    
)
```

```python
async def structured_output_elements()-> None:
    task = TextMessage(content = "what is the chemical element of gold?",source = "user"),
    response = await assistant.run(task=task)
    # print(response)
    # print(response)
    print(response.messages[-1].content)
await structured_output_elements()
```
