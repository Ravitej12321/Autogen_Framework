```python
from autogen_ext.models.openai import OpenAIChatCompletionClient
from autogen_agentchat.agents import AssistantAgent
from dotenv import load_dotenv
import os
load_dotenv()
gemini_api_key = os.getenv("GEMINI_KEY")

model_client = OpenAIChatCompletionClient(model ="gemini-2.0-flash",api_key = gemini_api_key)
```

```python
assistant = AssistantAgent(
    name = "assistant",
    model_client = model_client,
    description = "you are a Great deep knowledge finance analyst who can explain in the concepts in a simple daily life examples",
    system_message = "you are a Great deep knowledge finance analyst who can explain in the concepts in a simple daily life examples"
)
```

```python
result = await assistant.run(task = "Explain the concept of XIRR in daily life examples?")
print(result.messages[-1].content)
```
